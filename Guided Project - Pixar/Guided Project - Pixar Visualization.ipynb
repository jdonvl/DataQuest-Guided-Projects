Exploring a Pixar Movie Dataset

In this project, I aim to explore this dataset using the recently learned plotting techniques from Pandas, Seaborn, and Matplotlib. This dataset contains information on the first 15 Pixar movies including each movie's critics ratings, revenue figures, production costs, Oscar results, and more. 

### Introduction
The dataset has been loaded below.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

pixar_movies = pd.read_csv('PixarMovies.csv')
pixar_movies
```

### Data Cleaning
The last row describing the movie Inside Out is missing values for Oscars Nominated and Oscars Won. As per below, an error is generated and we're not able to perform calculations on those two columns. In below code, I create a new DataFrame excluding Inside Out.

Additionally, we can see that Domestic & International % are objects/strings, so I'll need to convert those into floats.

Last, I normalize the IMDB score to a 100 point scale and also add an Average Score (average of RT score, normalized IMDB score, and metacritic score) for later use.

```python
print(pixar_movies.shape)

print(pixar_movies.dtypes)

pixar_movies.describe()
```

```python
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str.rstrip('%').astype(float)
pixar_movies['International %'] = pixar_movies['International %'].str.rstrip('%').astype(float)
```

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score'] * 10
pixar_movies['Average Score'] = pixar_movies[['IMDB Score', 'RT Score', 'Metacritic Score']].mean(axis=1)
pixar_movies
filtered_pixar = pixar_movies.dropna()
filtered_pixar.set_index('Movie', inplace=True)
filtered_pixar
```

### Visuals

The below line chart compares how each movie fared across the 3 review sites. The immediate conclusion here is that Cars 2 fared the worst. Additionally, I can infer that RT generally reviews more favourably.

```python
critics_reviews = filtered_pixar[['RT Score', 'IMDB Score', 'Metacritic Score']]
critics_reviews.plot(figsize=(10,6), label=critics_reviews.index)
plt.xticks(np.arange(14),critics_reviews.index, rotation='vertical')
plt.show()
```

The box plot below confirms our hypothesis that Rotten Tomatoes appears to give consistently higher ratings. We can also see that Metacritic ratings for the Pixar movies are more spread out, ranging from right under 60 to right under 100.

```python
critics_reviews.plot(figsize=(9,5), kind='box')
plt.show()
```

The stacked bar plot below shows there's been a general decrease in the proportion of revenue that was made domestically.

```python
revenue_proportions = filtered_pixar[['Domestic %', 'International %']]
revenue_proportions.plot(figsize=(9,5), kind='bar', stacked=True)
plt.show()
```

The grouped bar plot below shows the Oscar Nominations and Oscars Won for each movie. A few of the movies won 2 Oscars, and Wall-E was the most nominated film of the bunch.

```python
oscars = filtered_pixar[['Oscars Nominated', 'Oscars Won']]
oscars.plot.bar()
plt.show()
```

I wanted to see if there was any relationship between the Adjusted Domestic Gross and Production Budget and Average Score so I created below pair plot. A pair plot will create a scatter plot for every combination of variables inputted. On the diagonal is a histogram for the individual variable. Based on below, with higher ratings, the movie has a higher domestic gross (top right chart).

```python
sns.pairplot(filtered_pixar[['Adjusted Domestic Gross','Production Budget','Average Score']])
sns.plt.show()
```
